```python
import os
os.getcwd()

# Load the dataset
df = pd.read_csv('aspirations.csv')

# Display the first 20 rows
print(df.head(20))

file_path = 'aspirations.csv'
data = pd.read_csv(file_path)
```

```python
# Assuming 'description' is the text data
texts = data['description'].values
labels = np.zeros(len(data))  # You can assign labels if needed

# Tokenization and Padding
max_words = 1000  # adjust as needed
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
```

```python
maxlen = 100  # adjust as needed
X = pad_sequences(sequences, maxlen=maxlen)

#  Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)
```

```python
model_rnn = Sequential()
model_rnn.add(Embedding(input_dim=max_words, output_dim=50, input_length=maxlen))
model_rnn.add(SimpleRNN(32))
model_rnn.add(Dense(1, activation='sigmoid'))

model_rnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_rnn.summary()
```

```python
#  Train RNN Model
model_rnn.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
#  Build LSTM Model
model_lstm = Sequential()
model_lstm.add(Embedding(input_dim=max_words, output_dim=50, input_length=maxlen))
model_lstm.add(LSTM(32))
model_lstm.add(Dense(1, activation='sigmoid'))

model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_lstm.summary()
```

```python
#  Train LSTM Model
model_lstm.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the RNN model
rnn_loss, rnn_accuracy = model_rnn.evaluate(X_test, y_test)

# Evaluate the LSTM model
lstm_loss, lstm_accuracy = model_lstm.evaluate(X_test, y_test)

# Print the results
print("RNN Model:")
print("- Loss:", rnn_loss)
print("- Accuracy:", rnn_accuracy)
print("\nLSTM Model:")
print("- Loss:", lstm_loss)
print("- Accuracy:", lstm_accuracy)
```
