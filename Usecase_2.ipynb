```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the CSV file
df = pd.read_csv('friend.csv')

# Display the first 20 rows
df.head(21)
```

## Finding duplicates in data

```python
print("Number of duplicate rows:", df.duplicated().sum())

for index, row in df.iterrows():
    name = row['Name']
    description = row['Description']
    if not description.startswith(name):
        df.at[index, 'Description'] = name + ": " + description
```

##Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense
```

##Load and Shuffle data

```python
df_shuffled = shuffle(df, random_state=42)
df_shuffled.head(21)
```

##Split df into Training and Testing Sets

```python
X = df_shuffled['Description'].values
y = df_shuffled['Name'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.33, random_state=42)
```

##Tokenize and Pad Sequences

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

max_sequence_length = max([len(seq) for seq in X_train_seq])
X_train_padded = pad_sequences(X_train_seq, maxlen=max_sequence_length, padding='post')
X_test_padded = pad_sequences(X_test_seq, maxlen=max_sequence_length, padding='post')
```

##Encode Labels


```python
label_to_index = {label: index for index, label in enumerate(np.unique(y_train))}
index_to_label = {index: label for label, index in label_to_index.items()}

# Assuming `y` is your full set of labels before splitting into training and testing sets
all_labels = np.concatenate([y_train, y_test])
unique_labels = np.unique(all_labels)

# Create a dictionary mapping each label to a unique index
label_to_index = {label: index for index, label in enumerate(unique_labels)}

# Now encode y_train and y_test using this dictionary
y_train_encoded = np.array([label_to_index[label] for label in y_train])
y_test_encoded = np.array([label_to_index[label] for label in y_test])
```

##Build and Compile RNN Model

```python
vocab_size = len(tokenizer.word_index) + 1  # Vocabulary size

model_rnn = Sequential([
    Embedding(input_dim=vocab_size, output_dim=100, input_length=max_sequence_length),
    SimpleRNN(units=64),
    Dense(units=len(label_to_index), activation='softmax')
])
model_rnn.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

##Build and Compile LSTM Model


```python
model_lstm = Sequential([
    Embedding(input_dim=vocab_size, output_dim=100, input_length=max_sequence_length),
    LSTM(units=64),
    Dense(units=len(label_to_index), activation='softmax')
])
model_lstm.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

##Train the Models

```python
# Train RNN model
model_rnn.fit(X_train_padded, y_train_encoded, validation_data=(X_test_padded, y_test_encoded), epochs=50, batch_size=64)

# Train LSTM model
model_lstm.fit(X_train_padded, y_train_encoded, validation_data=(X_test_padded, y_test_encoded), epochs=50, batch_size=64)
```

## Evaluate the Models

```python
# Evaluate RNN model
from sklearn.metrics import accuracy_score, classification_report

y_pred_rnn_probs = model_rnn.predict(X_test_padded)
y_pred_rnn = np.argmax(y_pred_rnn_probs, axis=1)
accuracy_rnn = accuracy_score(y_test_encoded, y_pred_rnn)
report_rnn = classification_report(y_test_encoded, y_pred_rnn)

print("RNN Model Accuracy:", accuracy_rnn)
print("RNN Model Classification Report:\n", report_rnn)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Now you can use accuracy_score and classification_report without any error
y_pred_lstm_probs = model_lstm.predict(X_test_padded)
y_pred_lstm = np.argmax(y_pred_lstm_probs, axis=1)
accuracy_lstm = accuracy_score(y_test_encoded, y_pred_lstm)
report_lstm = classification_report(y_test_encoded, y_pred_lstm)

print("LSTM Model Accuracy:", accuracy_lstm)
print("LSTM Model Classification Report:\n", report_lstm)
```
